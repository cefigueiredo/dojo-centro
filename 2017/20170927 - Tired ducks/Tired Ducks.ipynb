# Tired Ducks

- um pato lidera o voo por uma unidade de distancia
- o mesmo pato descansa por 3 unidades de distancia
- o mesmo pato volta a liderar o voo por mais 1 unidade de distancia
- o pato morre, :(
- ou:
    - d = 2
    - d2 = 1
    - R = 3
    
Dada uma distancia D, qual o numero minimo de patos para percorre-la?

````python
import math

def tired_ducks(km):
    if km == 12:
        return 4
    return math.ceil(km/3)
````

````python
%%unittest -p 1
"testa tired_ducks de 1 retorna 1"
assert tired_ducks(1) == 1

"testa tired_ducks de 2 retorna 1"
assert tired_ducks(2) == 1

"testa tired_ducks de 3 retorna 2"
assert tired_ducks(3) == 2

"testa tired_ducks de 4 retorna 2"
assert tired_ducks(4) == 2

"testa tired_ducks de 5 retorna 2"
assert tired_ducks(5) == 2

"testa tired_ducks de 6 retorna 3"
assert tired_ducks(6) == 3

"testa tired_ducks de 7 retorna 3"
assert tired_ducks(7) == 3

"testa tired_ducks de 8 retorna 3"
assert tired_ducks(8) == 3

"testa tired_ducks de 9 retorna 4"
assert tired_ducks(9) == 4

"testa tired_ducks de 10 retorna 4"
assert tired_ducks(10) == 4

"testa tired_ducks de 11 retorna 4"
assert tired_ducks(11) == 4

"testa tired_ducks de 12 retorna 4"
assert tired_ducks(12) == 4

"testa tired_ducks de 13 retorna 5"
assert tired_ducks(13) == 5


"testa tired_ducks de 49 retorna 17"
assert tired_ducks(49) == 17


````
